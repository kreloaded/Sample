# Iris Dataset Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("Iris.csv")
```

```python
data.head()
```

```python
print("Class Distribution : ")
data.groupby('iris').size()
```

```python
print("Statistical Analysis : ")
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
import seaborn as sns
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.violinplot(x="iris",y="sepal length",data=data)
plt.subplot(2,2,2)
sns.violinplot(x="iris",y="sepal width",data=data)
plt.subplot(2,2,3)
sns.violinplot(x="iris",y="petal length",data=data)
plt.subplot(2,2,4)
sns.violinplot(x="iris",y="petal width",data=data)
```

```python
sns.pairplot(data)
```

```python
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.boxplot(x="iris",y="sepal length",data=data)
plt.subplot(2,2,2)
sns.boxplot(x="iris",y="sepal width",data=data)
plt.subplot(2,2,3)
sns.boxplot(x="iris",y="petal length",data=data)
plt.subplot(2,2,4)
sns.boxplot(x="iris",y="petal width",data=data)
```
